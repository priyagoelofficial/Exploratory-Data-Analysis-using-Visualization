```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\Priya Goel\Downloads\Titanic.csv")
data.head()
```

```python
data['Age']=data['Age'].fillna(data['Age'].mean())
```

```python
data['Age']=data['Age'].round().astype('int64')
```

```python
data['Cabin']= data['Cabin'].fillna(data['Cabin'].mode()[0])
data['Embarked']= data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
data.head()
```

```python
data.duplicated().sum()
```

##### Q1. Create a countplot to show the number of passengers by gender.

```python
sns.countplot(data,x='Sex'
              ,color='green',
              width=0.5,
              edgecolor='red',
              linewidth=2)
plt.title("Count of Passengers by Gender")
plt.xlabel("Gender")
plt.ylabel("Count of Passengers")
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

Q2. Create a bar chart to show the number of passengers in each passenger class using different colors.

```python
pass_class=data.groupby("Pclass")['PassengerId'].count()

plt.Figure(figsize=(8,5))

plt.bar(pass_class.index,
        pass_class.values,
        color=['grey','yellow','pink'],
        edgecolor='blue',
        linewidth=2,
        width=0.6)

plt.title("Passenger in each Class",color='green',fontsize=14)
plt.xlabel("Class", color='orange')
plt.ylabel('Number of Passenger',color ='orange')
plt.grid(axis="y",linestyle="--",alpha=0.7)
plt.show()
        
```

Q3. Create a histogram to show the distribution of passenger age

```python
plt.Figure(figsize=(8,5))

plt.hist(data['Age'],
         bins=20,
         color='purple',
         edgecolor='blue')

plt.title('Distribution of Passenger Age',
          color="brown",
          fontsize=16)
plt.xlabel("Age",color="green",fontsize=9)
plt.ylabel("Frequency",color='green',fontsize=9)
plt.grid(axis='y',linestyle='--',alpha=0.5)
plt.show()
```

Q4. Create a boxplot to detect outliers in the fare column.

```python
sns.boxplot(y="Fare",data=data)
plt.show()
```

Q5. Create a scatterplot to 

 a. show the relationship between age and fare.

 b. analyze the relationship between family size (SibSp + Parch) and fare.
 

```python
sns.scatterplot(x='Fare',y='Age',data=data,
                size='Fare',
                color='Purple')
plt.title("Relationship between Age and Fare", color='brown', fontsize=18)
plt.xlabel('Fare',color='red', fontsize=12)
plt.ylabel('Age',color='red', fontsize=12)
plt.show()
```

```python
data['Family size']=data['SibSp']+data['Parch']
```

```python
sns.scatterplot(x='Family size',
                y="Fare", 
                data=data,
                color='orange',
                )
plt.title("Relationship between Family size and Fare",
          color='Red',
          fontsize=16)
plt.ylabel('Fare',color='brown',fontsize=10)
plt.xlabel('Family size',color='brown',fontsize=10)
plt.grid(axis='y',linestyle="--",alpha=0.7)
plt.show()
```

Q6. Create a countplot to show survival count by gender

```python
data['Survived_label']=data['Survived'].replace({
    0:"Die",
    1:"Living"
})
```

```python
a=data[data['Survived_label']=="Living"] 
```

```python
sns.countplot(x='Sex',
              hue='Survived_label',
              data=a,
            edgecolor='red',
            linewidth=2,
            width=0.5)
plt.ylabel('Survival Count',color='brown',fontsize=10)
plt.xlabel('Sex',color='brown',fontsize=10)
plt.grid(axis='y',linestyle="--",alpha=0.7)
plt.title("Survival count by Gnder",color='Green')
plt.show()
```

Q7. Create a bar chart to show survival by passenger class.

```python
data.head()
```

```python
survive_class=data.groupby('Pclass')['Survived'].sum()

plt.Figure(figsize=(8,5))

plt.bar(survive_class.index,
        survive_class.values,
        color=['Pink','Green','Blue'],
        width=0.5,
        linewidth=2,
        edgecolor='Black')
plt.title(' Survival By Passenger',color='Green',fontsize=18)
plt.xlabel("Passenger Class", color="orange",fontsize=14)
plt.ylabel("Survived", color="orange",fontsize=14)
plt.show()
```

Q8. Create a heatmap to show correlation between numerical variables.

```python
corr=data.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()
```
